```python
import pickle
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("df_w_embeddings_cluster.csv")
```

```python
result_np = df["embeddings"].apply(lambda x:[float(num) for num in x.strip('[]').split(',')])
result= np.array(result_np.to_list(), dtype=np.float64)
```

```python
num_clusters = 3

kmeans = KMeans(n_clusters = num_clusters, init='k-means++', max_iter=1500, n_init=40, random_state=0)

kmeans.fit(result)
```

```python
df["cluster"] = kmeans.labels_
```
