```python
import pandas as pd
import re
```

```python
df = pd.read_csv("df_w_embeddings_cluster_topiclabel_sentiment.csv")
```

```python
df["sentiments"]
```

```python
sentiments = []
for sentiment in df["sentiments"]:
    negative = sentiment[32:55]
    negative = re.sub('},', '', negative)
    negative = re.sub(r'[a-zA-Z]', '', negative)
    negative = negative.replace(r'[a-zA-Z]', "")
    negative = re.sub('}', '', negative)
    negative = re.sub('{', '', negative)
    negative = re.sub("'", '', negative)
    negative = float(negative)
    positive = sentiment[84:-2]
    positive = positive.replace(r'[a-zA-Z]', "")
    positive  = re.sub(r'[a-zA-Z]', '', positive )
    positive = re.sub(':', '', positive)
    positive = re.sub('r', '', positive)
    positive = re.sub('e', '', positive)
    positive = re.sub("'", '', positive)
    positive = re.sub("}", '', positive)
    positive = re.sub("}", '', positive)
    
    positive = float(positive)
    print('negative:', negative, 'positive:', positive)
    polarity = [-abs(negative), positive]
    print(sum(polarity))
    summ = sum(polarity)
    sentiments.append(summ)
```

```python
sentiments
```

```python
df["sentiments"] = sentiments
```

```python
df.to_csv("df_w_embeddings_cluster_topiclabel_sentiment_summed.csv")
```
